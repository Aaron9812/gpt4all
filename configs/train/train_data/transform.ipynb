```python
import pandas as pd
import json
```

```python
#load json file into df
df = pd.read_json('train_data.json',lines=True)
```

```python
df
```

```python
# create a json file that looks like this:
# {"prompt": f"Are these two product descriptions the same? Answer with yes or no", "response": if label 1 then "yes" else "no"}
# and save it as train_data_transformed.json
json_data = []  # List to store JSON objects

for index, row in df.iterrows():
    prompt = f"Are these two product descriptions the same? Answer with yes or no. Product1: {row['title_left']}, Product2: {row['title_right']}"
    response = "yes" if row['label'] == 1 else "no"
    
    json_object = {"prompt": prompt, "response": response}
    json_data.append(json_object)

# Save the JSON data to a file
with open("train_data_transformed.json", "w") as outfile:
    json.dump(json_data, outfile)
```
